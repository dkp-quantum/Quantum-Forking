```python
def qf1_rtest(axis,theta):
    # Create a Quantum Register with 3 qubit
    q = QuantumRegister(3)
    c = ClassicalRegister(1)

    # Create Quantum circuits for Testing simple QF
    q3 = QuantumCircuit(q,c)

    # Prepare initial state
    q3.h(q[0])
    if axis == 'x':
        q3.rx(theta,q[1])
    elif axis == 'y':
        q3.ry(theta,q[1])
    elif axis == 'z':
        q3.rz(theta,q[1])
    q3.barrier()
    q3.cswap(q[0],q[1],q[2])
    q3.barrier()
    q3.iden(q[1])
    q3.h(q[2])
    q3.barrier()
    q3.cswap(q[0],q[1],q[2])
    # Measure
    q3.measure(q[1], c[0])
    return q3
```
